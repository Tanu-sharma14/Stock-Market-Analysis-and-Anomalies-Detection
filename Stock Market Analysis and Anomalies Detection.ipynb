```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def fetch_stock_data(symbol, start_date, end_date):
    """
    Fetch historical stock data from Yahoo Finance.
    """
    data = yf.download(symbol, start=start_date, end=end_date)
    return data

def login():
    """
    Placeholder for login functionality.
    """
    username = input("Enter username: ")
    password = input("Enter password: ")
    # Check username and password
    if username == "Tanu" and password == "123":
        print("Login successful.")
        return True
    else:
        print("Login failed. Please try again.")
        return False

def update_profile():
    """
    Placeholder for updating profile functionality.
    """
    print("Update profile functionality placeholder")

def plot_stock_data(data, title):
    """
    Plot stock data using matplotlib.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

def compare_stocks(stock_data1, stock_data2, ticker1, ticker2):
    """
    Compare stock data between two tickers and display the comparison.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data1['Close'], label=f'{ticker1} Close Price', color='blue')
    plt.plot(stock_data2['Close'], label=f'{ticker2} Close Price', color='green')
    plt.title('Stock Price Comparison')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Calculate profit comparison
    profit1 = stock_data1['Close'].iloc[-1] - stock_data1['Close'].iloc[0]
    profit2 = stock_data2['Close'].iloc[-1] - stock_data2['Close'].iloc[0]

    # Create a comparison table
    comparison_data = {
        'Ticker': [ticker1, ticker2],
        'Profit': [profit1, profit2]
    }
    comparison_df = pd.DataFrame(comparison_data)
    print("\nProfit Comparison:")
    print(comparison_df)

    # Update profile if user is logged in
    update_profile()

def high_frequency_trading_strategies(stock_data1, stock_data2, ticker1, ticker2):
    """
    Implement high-frequency trading strategies for the inputted stock tickers.
    """
    short_window = 5
    long_window = 20

    # Calculate moving averages
    stock_data1['Short_MA'] = stock_data1['Close'].rolling(window=short_window).mean()
    stock_data1['Long_MA'] = stock_data1['Close'].rolling(window=long_window).mean()

    stock_data2['Short_MA'] = stock_data2['Close'].rolling(window=short_window).mean()
    stock_data2['Long_MA'] = stock_data2['Close'].rolling(window=long_window).mean()

    # Generate buy/sell signals based on the crossover
    stock_data1['Signal'] = np.where(stock_data1['Short_MA'] > stock_data1['Long_MA'], 1, 0)
    stock_data1['Position'] = stock_data1['Signal'].diff()

    stock_data2['Signal'] = np.where(stock_data2['Short_MA'] > stock_data2['Long_MA'], 1, 0)
    stock_data2['Position'] = stock_data2['Signal'].diff()

    # Plot the moving averages and the signals
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data1.index, stock_data1['Close'], label=f'{ticker1} Close Price', color='blue')
    plt.plot(stock_data1.index, stock_data1['Short_MA'], label='Short MA', color='green')
    plt.plot(stock_data1.index, stock_data1['Long_MA'], label='Long MA', color='red')
    plt.plot(stock_data1.index, stock_data1['Position'] * 100, label='Position', marker='o', linestyle='', color='purple')
    plt.title('High-Frequency Trading Strategy')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Print the positions
    print(f"\nTrading Signals for {ticker1}:")
    print(stock_data1[['Short_MA', 'Long_MA', 'Position']].tail())

    plt.figure(figsize=(12, 6))
    plt.plot(stock_data2.index, stock_data2['Close'], label=f'{ticker2} Close Price', color='blue')
    plt.plot(stock_data2.index, stock_data2['Short_MA'], label='Short MA', color='green')
    plt.plot(stock_data2.index, stock_data2['Long_MA'], label='Long MA', color='red')
    plt.plot(stock_data2.index, stock_data2['Position'] * 100, label='Position', marker='o', linestyle='', color='purple')
    plt.title('High-Frequency Trading Strategy')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Print the positions
    print(f"\nTrading Signals for {ticker2}:")
    print(stock_data2[['Short_MA', 'Long_MA', 'Position']].tail())

    # Update profile if user is logged in
    update_profile()

def portfolio_optimization(stock_data1, stock_data2, ticker1, ticker2):
    """
    Implement portfolio optimization for the inputted stock tickers.
    """
    # Combine the stock data into a single DataFrame
    combined_data = pd.concat([stock_data1['Close'], stock_data2['Close']], axis=1)
    combined_data.columns = [ticker1, ticker2]

    # Calculate daily returns
    daily_returns = combined_data.pct_change().dropna()

    # Calculate expected returns and covariance matrix
    expected_returns = daily_returns.mean()
    cov_matrix = daily_returns.cov()

    # Set the number of simulations
    num_portfolios = 10000

    # Initialize results arrays
    results = np.zeros((3, num_portfolios))
    weights_record = []

    for i in range(num_portfolios):
        weights = np.random.random(2)
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_return = np.sum(weights * expected_returns) * 252
        portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
        results[0,i] = portfolio_return
        results[1,i] = portfolio_std_dev
        results[2,i] = results[0,i] / results[1,i]

    # Find the optimal portfolio
    max_sharpe_idx = np.argmax(results[2])
    optimal_portfolio_return = results[0,max_sharpe_idx]
    optimal_portfolio_std_dev = results[1,max_sharpe_idx]
    optimal_portfolio_weights = weights_record[max_sharpe_idx]

    # Plot the efficient frontier
    plt.figure(figsize=(12, 6))
    plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis', marker='o')
    plt.title('Efficient Frontier')
    plt.xlabel('Volatility')
    plt.ylabel('Return')
    plt.colorbar(label='Sharpe Ratio')
    plt.scatter(optimal_portfolio_std_dev, optimal_portfolio_return, marker='x', color='r', s=200, label='Optimal Portfolio')
    plt.legend()
    plt.show()

    # Print the optimal portfolio weights
    print("\nOptimal Portfolio Weights:")
    print(f"{ticker1}: {optimal_portfolio_weights[0]:.2f}")
    print(f"{ticker2}: {optimal_portfolio_weights[1]:.2f}")

    # Update profile if user is logged in
    update_profile()

def detect_anomalies(stock_data, ticker):
    """
    Detect anomalies in stock market data.
    """
    # Calculate daily returns
    daily_returns = stock_data['Close'].pct_change().dropna()

    # Calculate mean and standard deviation
    mean_return = daily_returns.mean()
    std_dev = daily_returns.std()

    # Define anomaly threshold (e.g., 3 standard deviations from the mean)
    anomaly_threshold = mean_return + 3 * std_dev

    # Find dates with anomalies
    anomalies = daily_returns[daily_returns > anomaly_threshold]

    # Plot anomalies
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label=f'{ticker} Close Price', color='blue')
    plt.plot(anomalies.index, stock_data.loc[anomalies.index]['Close'], 'ro', label='Anomalies')
    plt.title('Anomalies Detection')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Print anomalies
    print("\nAnomalies Detected:")
    print(anomalies)

    # Update profile if user is logged in
    update_profile()

def main():
    logged_in = False
    while True:
        print("\nMenu:")
        print("1. Login")
        if logged_in:
            print("2. Compare Stocks")
            print("3. Portfolio Optimization")
            print("4. High-Frequency Trading Strategies")
            print("5. Market Anomaly Detection")
            print("6. Logout")
        else:
            print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1' and not logged_in:
            logged_in = login()
        elif choice == '6':
            if logged_in:
                print("You are logged out.")
                logged_in = False
            else:
                print("Exiting program.")
                break
        elif not logged_in:
            print("Please login first.")
        elif choice == '2' and logged_in:
            ticker1 = input("Enter first stock symbol: ")
            ticker2 = input("Enter second stock symbol: ")
            start_date = input("Enter start date (YYYY-MM-DD): ")
            end_date = input("Enter end date (YYYY-MM-DD): ")
            if start_date >= end_date:
                print("Error: Start date must be before end date.")
                continue
            stock_data1 = fetch_stock_data(ticker1, start_date, end_date)
            stock_data2 = fetch_stock_data(ticker2, start_date, end_date)

            # Display data for the first ticker
            print(f"\nActual Stock Data for {ticker1}:")
            print(stock_data1)
            plot_stock_data(stock_data1, f"Stock Price for {ticker1}")

            # Display data for the second ticker
            print(f"\nActual Stock Data for {ticker2}:")
            print(stock_data2)
            plot_stock_data(stock_data2, f"Stock Price for {ticker2}")

            # Compare stock data and display the comparison
            compare_stocks(stock_data1, stock_data2, ticker1, ticker2)
        elif choice == '3' and logged_in:
            ticker1 = input("Enter first stock symbol: ")
            ticker2 = input("Enter second stock symbol: ")
            start_date = input("Enter start date (YYYY-MM-DD): ")
            end_date = input("Enter end date (YYYY-MM-DD): ")
            if start_date >= end_date:
                print("Error: Start date must be before end date.")
                continue
            stock_data1 = fetch_stock_data(ticker1, start_date, end_date)
            stock_data2 = fetch_stock_data(ticker2, start_date, end_date)
            portfolio_optimization(stock_data1, stock_data2, ticker1, ticker2)
        elif choice == '4' and logged_in:
            ticker1 = input("Enter first stock symbol: ")
            ticker2 = input("Enter second stock symbol: ")
            start_date = input("Enter start date (YYYY-MM-DD): ")
            end_date = input("Enter end date (YYYY-MM-DD): ")
            if start_date >= end_date:
                print("Error: Start date must be before end date.")
                continue
            stock_data1 = fetch_stock_data(ticker1, start_date, end_date)
            stock_data2 = fetch_stock_data(ticker2, start_date, end_date)
            high_frequency_trading_strategies(stock_data1, stock_data2, ticker1, ticker2)
        elif choice == '5' and logged_in:
            ticker = input("Enter stock symbol: ")
            start_date = input("Enter start date (YYYY-MM-DD): ")
            end_date = input("Enter end date (YYYY-MM-DD): ")
            if start_date >= end_date:
                print("Error: Start date must be before end date.")
                continue
            stock_data = fetch_stock_data(ticker, start_date, end_date)
            detect_anomalies(stock_data, ticker)
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
